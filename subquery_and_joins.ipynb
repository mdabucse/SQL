### EXISTS
* Checks whether a subquery returns at least one row and returns TRUE or FALSE based on that
* oru column irunthalum fulla a antha table a return pannum apdi illaina ethayume return pannathu

```python
%%sql
select customer_name 
from customers
where exists (
    select * from customers where city = 'chennai'
                            );
```

```python
%%sql
-- instead of joins we use subquery for small tables
select 
    customer_name,
    city,
    (select sum(order_amount)
    from orders o  
    where o.customer_id = c.customer_id) as total_orders
from customers c;
```

```python
%%sql
-- this alice is used for an temperory table it wont stored in the system
select customer_name from
(select * from customers) as name;
```

```python
%%sql
-- in case statement
select 
    customer_name,
    case
        when (select sum(order_amount) from orders where orders.customer_id=customers.customer_id)>
             (select avg(order_amount) from orders) then 'Above Average'
        else 'Below Average'
    end as order_category
from customers
```

```python
%%sql
drop table if exists orders ;
CREATE TABLE orders (
    order_id INT PRIMARY KEY,
    restaurant_id INT,
    order_date DATE
)
```

### Inner join or Join
* it takes only both of the table commen content alone

```python
%%sql
select r.name,o.order_date
from Restaurants r 
join orders o on r.id = o.restaurant_id
```

### Left Join
* it takes both of the table commen content and left side content

```python
%%sql
select r.name,o.order_date
from Restaurants r 
left join orders o on r.id = o.restaurant_id
```

### Inner Right Join
* it takes both of the table commen content and right side content

```python
%%sql
select r.name,o.order_date
from Restaurants r 
right join orders o on r.id = o.restaurant_id
```

## Self Join
* to join the same table itself

```python
%%sql
DROP TABLE IF EXISTS Employees;

CREATE TABLE Employees (
    id INT PRIMARY KEY,
    name VARCHAR(100) NOT NULL,
    manager_id INT NULL
);

INSERT INTO Employees (id, name, manager_id) VALUES
(1, 'Alice', 3),
(2, 'Bob', 3),
(3, 'Charlie', NULL),
(4, 'Diana', 3);
```

```python
%%sql
select *
from Employees e
left join Employees m
on e.manager_id = m.id;
```

```python
%%sql
select 
    e.name,
    m.name
from Employees e
left join Employees m
on e.manager_id = m.id;
```
