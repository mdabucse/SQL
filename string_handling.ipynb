### String Handling

```python
%%sql
use uber
```

```python
%%sql
show tables
```

```python
%%sql
select * from customerdata 
```

```python
%%sql
-- Length of string in the customer data name 
select length(customer_name) as length from customerdata
```

```python
%%sql
select upper(address) as address_name_in_upper from customerdata
```

```python
%%sql
select lower(address) as address_name_in_lower from customerdata
```

```python
%%sql
select concat(customer_name," The city of  ",address) as name_with_address from customerdata
```

```python
%%sql
-- (value , start, end ) params for substring
select substring(address,1,3) as sub from customerdata
```

```python
%%sql
-- remove all white spaces from the string 
select trim("   abu   ")
```

```python
%%sql
-- padding right
select rpad("abu",15,'*');
```

```python
%%sql
-- padding left
select lpad("abu",15,'*')
```

```python
%%sql
select replace(customer_name,'a','-') as updated_name from customerdata
```

```python
%%sql
-- to see the position of some character
select instr(customer_name,'a') as position_of_a from customerdata
```

```python
%%sql
select left(customer_name,3) as first_3_characters from customerdata
```

```python
%%sql
select right(customer_name,3) as last_3_characters from customerdata
```

```python
%%sql
select reverse(customer_name) as reversed_name from customerdata;
```

```python
%%sql
select format(987654321,2) as format_number
```

```python
%%sql
-- nested functions usage
select upper(concat(customer_name," The city of  ",address)) as name_with_address from customerdata
```
